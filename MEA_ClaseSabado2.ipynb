```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import statsmodels.api as sm
# importing train_test_split from sklearn
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("D:\MaestriaEA\Metodos estadísticos avanzados\clase3\MATRICULA_OFICIAL_EN_LOS_MUNICIPIOS_DEL_MAGDALENA__SEGUN_ZONA__JORNADA_E_INSTITUCIONES_EDUCATIVAS_EN_LOS_A_OS_A_OS_2020_Y_2021.csv",delimiter=";")
```

```python
NonIns=(data["NOMBRE INSTITUCION"].unique())
data0=data.copy()
```

```python
data0["NonIns"]=data["NOMBRE INSTITUCION"]
```

```python
data0["HORARIO"].unique()
```

```python
len(data["NOMBRE INSTITUCION"])
```

```python
data0["Thorario"]=data0["HORARIO"]
```

```python
data0["Thorario"]=data0["Thorario"].replace("DIURNO",1)
```

```python
j=1
for i in NonIns:
    data0["NonIns"]=data0["NonIns"].replace(i,j)
    j+=1
```

```python
data0=data0.drop(columns=["AÑO"])
```

```python
Y = data0["NonIns"]
X = data0.drop(columns=["NonIns"])
X = sm.add_constant(X)
```

```python
X
```

```python
model_linear = sm.OLS(Y, X).fit()
predictions_SM = model_linear.predict(X) 
```

```python
model_linear.summary()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data0["NonIns"],predictions_SM)
u=np.linspace(data0["NonIns"].min(),data0["NonIns"].max(),5)
plt.plot(u,u)
plt.xlabel("Nombre de Instituciones")
plt.ylabel("Predicciones")
plt.show()
```

```python
rmse=np.sqrt(np.mean((data0["NonIns"]-predictions_SM)**2))
print("RMSE del modelo es",rmse)
```

Ahora probemos con machine learning a hacer regresiones lineales!

```python
features=data0.drop(columns=["NonIns"])
target=data0["NonIns"]
# splitting the data
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state = 746)
```

```python
# importing module
from sklearn.linear_model import LinearRegression
# creating an object of LinearRegression class
LR = LinearRegression()
# fitting the training data
LR.fit(x_train,y_train)
prediction_LR =  LR.predict(x_test)
prediction_LR
```

```python
# get importance
var=LR.feature_names_in_
importance = LR.coef_
# summarize feature importance
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
plt.barh(var, importance)
plt.show()
```

Hagamos ahora una nueva prueba solo usando las variables que me hemos conseguido importantes:

```python
predictions_LR=LR.predict(features)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data0["NonIns"],predictions_SM, label="Statsmodels",alpha=0.5)
plt.scatter(data0["NonIns"],predictions_LR, label="Machine Learning",s=5,c='r')
u=np.linspace(data0["NonIns"].min(),data0["NonIns"].max(),5)
plt.plot(u,u)
plt.xlabel("NOMBRE DE INSTITUCIONES")
plt.ylabel("PREDICCIONES")
plt.legend()
plt.show()
```

```python
LR2 = LinearRegression(fit_intercept=False)
# fitting the training data
LR2.fit(x_train,y_train)
predictions_LR2=LR2.predict(features)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data0["NonIns"],predictions_SM, label="Statsmodels with intercept")
plt.scatter(data0["NonIns"],predictions_LR, label="Machine Learning with intercept")
plt.scatter(data0["NonIns"],predictions_LR2, label="Machine Learning without intercept")
u=np.linspace(data0["NonIns"].min(),data0["NonIns"].max(),5)
plt.plot(u,u)
plt.xlabel("NOMBRE DE INSTITUCIONES")
plt.ylabel("PREDICCIONES")
plt.legend()
plt.show()
```

```python
# importing r2_score module
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

# predicting the accuracy score
score_SM=r2_score(target,predictions_SM)
score_LR=r2_score(target,predictions_LR)
score_LR2=r2_score(target,predictions_LR2)
RMS2_SM=np.sqrt(mean_squared_error(target,predictions_SM))
RMS2_LR=np.sqrt(mean_squared_error(target,predictions_LR))
RMS2_LR2=np.sqrt(mean_squared_error(target,predictions_LR2))

for i,j in zip([score_SM,score_LR,score_LR2],["score_SM","score_LR","score_LR2"]):
    print(j," is ",i)
for i,j in zip([RMS2_SM,RMS2_LR,RMS2_LR2],["RMS2_SM","RMS2_LR","RMS2_LR2"]):
    print(j," is ",i)
```

**Ahora regresión con Random Forest!!!!!**

```python
# Fitting Random Forest Regression to the dataset
# import the regressor
from sklearn.ensemble import RandomForestRegressor
 
 # create regressor object
regressor = RandomForestRegressor(n_estimators = 100, random_state = 746)
 
# fit the regressor with x and y data
regressor.fit(features, target) 
predictions_RF=regressor.predict(features)
```

```python
score_RF=r2_score(target,predictions_RF)
RMS2_RF=np.sqrt(mean_squared_error(target,predictions_RF))
```

```python
for i,j in zip([score_SM,score_LR,score_LR2,score_RF],["score_SM","score_LR","score_LR2","score_RF"]):
    print(j," \tis ",i)
for i,j in zip([RMS2_SM,RMS2_LR,RMS2_LR2,RMS2_RF],["RMS2_SM","RMS2_LR","RMS2_LR2","RMS2_RF"]):
    print(j," \tis ",i)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data0["NonIns"],predictions_SM, label="Statsmodels with intercept",alpha=0.5)
plt.scatter(data0["NonIns"],predictions_LR, label="Machine Learning with intercept",s=5)
plt.scatter(data0["NonIns"],predictions_LR2, label="Machine Learning without intercept",s=4)
plt.scatter(data0["NonIns"],predictions_RF, label="Machine Learning Random Forest",s=7)
u=np.linspace(data0["NonIns"].min(),data0["NonIns"].max(),5)
plt.plot(u,u)
plt.xlabel("NOMBRE DE iNSTITUCIONES")
plt.ylabel("PREDICCIONES")
plt.legend()
plt.show()
```
